# PUI 2018 Extra Credit Project

# Background

New York City water is virtually lead-free when it is delivered from the City’s upstate reservoir
system, but water can absorb lead from solder, fixtures, and pipes found in the plumbing of some
buildings or homes.

High lead levels in water are often associated with learning and development disability in children.


We seek to establish a relationship between lead water levels in drinking water of the schools and
the performance of the children of those schools across various age groups, gender and race.

# References:

https://www.health.ny.gov/press/releases/2017/docs/lead_in_school_drinking_water_report.pdf


https://www.nytimes.com/2017/04/28/nyregion/new-york-schools-lead-water.html


# Objective

 Analysis to check if the presence of lead > 15ppb in the water outlets of the schools in New York City are an indicator of the overall mean scale score for Mathematics for the grades 3-8.

# Deliverables

This analysis spans 4 levels:

1. Observation of correlation between the high lead levels and the mean scale score for Mathematics.

2. Test for the gaussianity of the distribution of the mean scale score for Mathematics.

3. Test for the difference in the proportion of the males and females having a mean scale score than the average mean scale score for Mathematics for the schools having the presence of lead > 15ppb in the water outlets.

4. Test for the difference in the proportion of the various ethnicities(Asian, Hispanic, Black, White) having a mean scale score than the average mean scale score for Mathematics for the schools having the presence of lead > 15ppb in the water outlets.

5. Test for the difference in the proportion of the different economic backgrounds(Economically Disadvantaged and Not Economically Disadvantaged)having a mean scale score than the average mean scale score for Mathematics for the schools having the presence of lead > 15ppb in the water outlets.

# Correlation

#### $H_{0}$ : There is no correlation between the lead levels in the water of the schools and the mean scale score in Mathematics  for all grades of that school. Significance level {alpha} = 0.05

#### $H{a}$: There is a negative correlation between the lead levels in the water of the schools and the mean scale score in Mathematics  for all grades of that school.

```python
pearson_values = stats.pearsonr(high_lead_levels_grades['Mean Scale Score'].values ,high_lead_levels_grades['Number of Outlets, Result > 15 ppb'].values ) 
```

```python
pearson_values
```

# Conclusion:

* There appears to be a slight negative correlation, ie higher lead levels result in lower grades and vice versa.

* However , the high p-value indicates that there is a high probability that the result was possibly observed by chance. 


Therefore, we fail to reject the null hypothesis that there is no correlation between the lead levels in the water of the schools and the mean scale score in Mathematics for all grades of that school

### $H_{0}$ : There is no difference in the proportion of boys of the schools with mean scale score greater than the average score and the proportion of girls with mean scale score greater than the average score of the schools having > 0 high lead water levels .

### $H_{1}$ : There is a difference in the proportion of boys of the schools with mean scale score greater than the average score and the proportion of girls with mean scale score greater than the average score of the schools having > 0 high lead water levels .

```python
male_female_chisq = stats.chisquare([(nyc_schools_maths_grades_gender_2017_males_high_lead_levels['Mean Scale Score'] \
                                      > nyc_schools_maths_grades_gender_2017_males['Mean Scale Score'].mean()).sum(),
(nyc_schools_maths_grades_gender_2017_females_high_lead_levels['Mean Scale Score'] > \
 nyc_schools_maths_grades_gender_2017_females['Mean Scale Score'].mean()).sum()] )

if male_female_chisq.pvalue < 0.05:
    print('Reject the null hypothesis')
else:
    print('Fail to reject the null hypothesis')
    print(male_female_chisq)
```

# Chi-Square Test for Homogeneity


### $H_{0}$ : There is no difference in the proportion of the various ethnicities of the schools having >15ppb high lead level in water outlets with mean scale score greater than the average mean scale score for these schools.

### $H_{1}$ : There is a difference in the proportion of the various ethnicities of the schools having >15ppb high lead level in water outlets with mean scale score greater than the average mean scale score for these schools.

# Chi-Square Test for Homogeneity


### $H_{0}$ : There is no difference in the proportion of the economic backgrounds(ie economically disadvantaged and NOT economically disadvantaged) of the schools having > 15ppb  lead water levels with mean scale score greater than the average mean scale score of all such schools.

### $H_{1}$ : There is a difference in the proportion of the economic backgrounds of the schools having > 15ppb high lead water levels with mean scale score greater than the average mean scale score of all such schools.

### Significance level : alpha = 0.05


```python
eco_chisq = stats.chisquare([(nyc_schools_maths_grades_econadv_highlead_2017['Mean Scale Score'] > nyc_schools_maths_grades_econ_2017['Mean Scale Score'].mean()).sum(),
(nyc_schools_maths_grades_econdisadv_highlead_2017['Mean Scale Score'] > nyc_schools_maths_grades_econ_2017['Mean Scale Score'].mean()).sum()
])
if eco_chisq.pvalue < 0.05:
    print('Reject the null')
    print(eco_chisq)
else:
    print('Fail to reject the null')
    
```

```python
schools_shp_high_lead_levels_econadv = schools_shp.merge(nyc_schools_maths_grades_econadv_highlead_2017,left_on='LOC_CODE' , right_on='School')
schools_shp_high_lead_levels_econdisadv = schools_shp.merge(nyc_schools_maths_grades_econdisadv_highlead_2017,left_on='LOC_CODE' , right_on='School')
```
