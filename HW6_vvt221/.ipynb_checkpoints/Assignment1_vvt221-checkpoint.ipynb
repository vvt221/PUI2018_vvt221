# Use groupby to count the building per LandUse code

## Manhattan

```python
mh_glu = manhattan.groupby(by=['LandUse']).count().reset_index()
mh_glu
```

## Brooklyn

```python
bk_glu = brooklyn.groupby(by=['LandUse']).count().reset_index()
bk_glu
```

# Convert the landuse to integer

## Manhattan

```python
mh_glu['LandUse'] = mh_glu['LandUse'].astype('int64')
mh_glu.dtypes
```

## Brooklyn

```python
bk_glu['LandUse'] = bk_glu['LandUse'].astype('int64')
bk_glu.dtypes
```

# Merge the dataframes on land use

```python
bkmh = bk_glu.merge(mh_glu ,how='outer', on='LandUse')
bkmh
```

```python
bkmh.columns
```

# Rename things so that you have descriptive names

```python
bkmh.rename(columns={'ZipCode_x':'bk','ZipCode_y':'mh'},inplace=True)
bkmh
```

# Find online the LandUse data dictionary and replace code w the verbose description of it

```python
md = {1: 'One &Two Family Buildings',
      2: 'Multi-Family Walk-Up Buildings',
      3: 'Multi-Family Elevator Buildings', 
      4: 'Mixed Residential & Commercial Buildings',
      5: 'Commercial & Office Buildings',
      6: 'Industrial & Manufacturing', 
      7: 'Transportation & Utility',
      8: 'Public Facilities & Institutions',
      9: 'Open Space & Outdoor Recreation',
      10: 'Parking Facilities',
      11: 'Vacant Land'}
bkmh["LandUse"].replace(md,inplace=True)

bkmh
```

# Create a feature (a column, a series) with the ratio of number of buildings in BK vs MN and find out in which LandUse class Manhattan has more buildings than BK

```python
bkmh['luratio'] = bkmh['mh'] / bkmh['bk']
bkmh
```

```python
luratio = bkmh['mh'] / bkmh['bk']
luratio
```

```python
def ratio_check(inputSeries):
    if inputSeries > 1:
        return True
    else:
        return False
luratio.apply(ratio_check)        
```

```python
bkmh['Manhattan wins'] = luratio.apply(ratio_check)   
```
